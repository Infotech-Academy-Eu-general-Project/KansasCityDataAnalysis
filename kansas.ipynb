#### Replace spaces with underscore , lowercase labels and rename columns

```python
def strip_special_char(df):
    
    df.rename(columns=lambda x : x.strip().replace(':@',''), inplace=True)
    df.rename(columns= lambda x: x.strip().lower().replace(' ','_'), inplace=True)
    df.rename(columns= lambda x: x.strip().lower().replace(':',''), inplace=True)
    return df

df_18 = strip_special_char(df_18)
df_19 = strip_special_char(df_19)
df_20 = strip_special_char(df_20)
df_21 = strip_special_char(df_21)
df_22 = strip_special_char(df_22)    
```

```python
def assign_dtype(df):

    df.firearm_used_flag=df.firearm_used_flag.astype(bool)
    #####################################################################
    df.computed_region_9t2m_phkm=df.computed_region_9t2m_phkm.astype(float)
    df.computed_region_kk66_ngf4=df.computed_region_kk66_ngf4.astype(float)
    df.computed_region_my34_vmp8=df.computed_region_my34_vmp8.astype(float)
    df.computed_region_w4hf_t6bp=df.computed_region_w4hf_t6bp.astype(float)
    #####################################################################
    df.reported_date=pd.to_datetime(df.reported_date, infer_datetime_format=True)
    df.from_date=pd.to_datetime(df.from_date, infer_datetime_format=True)
    df.to_date=pd.to_datetime(df.to_date, infer_datetime_format=True)
    #####################################################################
    df.reported_time = pd.to_datetime(df.reported_time, format='%H:%M').dt.strftime("%H:%M")
    df.from_time = pd.to_datetime(df.from_time, format='%H:%M').dt.strftime("%H:%M") 
    df.to_time = pd.to_datetime(df.to_time, format='%H:%M').dt.strftime("%H:%M")

    return df
```
